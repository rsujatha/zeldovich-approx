```python
import numpy as np
import copy
import matplotlib.pyplot as plt
import gc
import time
```

## Inverse Fourier Transform using Numpy irfft

```python
## Setting Variance of the random numbers
GridSize = 256
dk=0.3
kx = np.concatenate([range(0,GridSize/2),range(-GridSize/2,0)])*dk
kxx,kyy = np.meshgrid(kx,kx[0:GridSize/2+1], sparse=True,indexing='ij')
k = kxx**2+kyy**2
variance = np.ones([GridSize,GridSize/2+1])*k*np.exp(-k**2)
mean = 0
```

## Below the numpys irfft2 is tested against ifft2 
#### Three 2d arrays are used below:                                                                                                                       
#### 1) deltak_half is random nuber array of size [GridSize,GridSize/2+1]                                                       
#### 2) deltak_half_processed is the copy of deltak_half but the end points are modified to make ift real                                     
#### 3) deltak is the fullsized version of deltak_half_processed including both +ve and -ve frequencies

### Here irfft is performed on deltak_half and deltak_half_processed and compared with ifft on deltak

```python
deltak_half 									= np.random.normal(mean,np.sqrt(variance), [GridSize,GridSize/2+1]) + np.random.normal(mean,np.sqrt(variance), [GridSize,GridSize/2+1])*1j

### Making ifft real manually
deltak 									= np.empty([GridSize,GridSize])*1j
deltak[:,0:GridSize/2+1] 				= deltak_half
deltak[0,0]								= 0
deltak[0,GridSize/2]					= deltak[0,GridSize/2].real
deltak[GridSize/2,0]					= deltak[GridSize/2,0].real
deltak[GridSize/2,GridSize/2]			= deltak[GridSize/2,GridSize/2].real
deltak[GridSize/2+1:,0]					= deltak[1:GridSize/2,0][::-1].conjugate()
deltak[0,GridSize/2+1:]					= deltak[0,1:GridSize/2][::-1].conjugate() 
deltak[GridSize/2+1:,GridSize/2]		= deltak[1:GridSize/2,GridSize/2][::-1].conjugate()
deltak[1:GridSize,GridSize/2+1:] = deltak[1:GridSize,1:GridSize/2][::-1,::-1].conjugate()

deltak_half_processed = deltak[:,0:GridSize/2+1]
```

```python
### inverse fourier transform
deltax_half = np.fft.irfft2(deltak_half)
deltax_half_processed = np.fft.irfft2(deltak_half_processed)
deltax = np.fft.ifft2(deltak)
```

```python
fig = plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.title('deltax_half')
plt.pcolormesh(deltax_half)
plt.colorbar()
plt.subplot(1,3,2)
plt.pcolormesh(deltax.real)
plt.title('deltax')
plt.colorbar()
plt.subplot(1,3,3)
plt.pcolormesh(deltax_half_processed)
plt.title('deltax_half_processed')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
hist,edge = np.histogram(deltax_half_processed.flatten()-deltax.real.flatten(),bins=20)
plt.plot((edge[1:]+edge[:-1])/2,hist)
plt.title('histogram')
plt.ylabel('counts')
plt.xlabel('deltax_half_processed - deltax')
plt.subplot(1,2,2)
hist,edge = np.histogram(deltax_half_processed.flatten()/deltax.real.flatten()-1,bins=20)
plt.plot((edge[1:]+edge[:-1])/2,hist)
plt.title('histogram')
plt.ylabel('counts')
plt.xlabel('error%')
plt.show()
fig = plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.locator_params(nbins=4)
hist,edge = np.histogram(deltax_half.flatten()-deltax.real.flatten(),bins=20)
plt.plot((edge[1:]+edge[:-1])/2,hist)
plt.title('histogram')
plt.ylabel('counts')
plt.xlabel('deltax_half - deltax')
plt.subplot(1,2,2)
hist,edge = np.histogram(deltax_half.flatten()/deltax.real.flatten()-1,bins=20)
plt.plot((edge[1:]+edge[:-1])/2,hist)
plt.title('histogram')
plt.ylabel('counts')
plt.xlabel('error ')
plt.show()
```

```python
del deltak,deltak_half,deltax_half_processed
gc.collect()
```

## Below the numpys irfftn is tested against ifftn (for n=3 ) 
#### Three 2d arrays are used below:                                                                                                                       
#### 1) deltak_half is random nuber array of size [GridSize,GridSize/2+1]                                                       
#### 2) deltak_half_processed is the copy of deltak_half but the end points are modified to make ift real                                     
#### 3) deltakf is the fullsized version of deltak_half_processed including both +ve and -ve frequencies

### Here irfft is performed on deltak_half and deltak_half_processed and compared with ifft on deltak

```python
GridSize = 256
dk=0.2
kx = np.concatenate([range(0,GridSize/2),range(-GridSize/2,0)])*dk
kxx,kyy,kzz = np.meshgrid(kx,kx,kx[0:GridSize/2+1], sparse=True,indexing='ij')
k = kxx**2+kyy**2+kzz**2
variance = 10*np.ones([GridSize,GridSize/2+1])*k*np.exp(-k**2)
mean = 0
```

```python
deltak_half 									= np.random.normal(mean,np.sqrt(variance), [GridSize,GridSize,GridSize/2+1]) + np.random.normal(mean,np.sqrt(variance), [GridSize,GridSize,GridSize/2+1])*1j

deltakf = np.empty([GridSize,GridSize,GridSize])*1j
deltakf[:,:,0:GridSize/2+1] = deltak_half
deltakf[1:,1:,GridSize/2+1:] = deltakf[1:,1:,1:GridSize/2][::-1,::-1,::-1].conjugate()
deltakf[0,0,0]=0
deltakf[0,0,GridSize/2] = deltakf[0,0,GridSize/2].real*np.sqrt(2) ## this multiplication by sqrt(2) is since the random number is purely real the variance must be twice
deltakf[GridSize/2,0,0] = deltakf[GridSize/2,0,0].real*np.sqrt(2)
deltakf[0,GridSize/2,0] = deltakf[0,GridSize/2,0].real*np.sqrt(2)
deltakf[0,GridSize/2,GridSize/2] = deltakf[0,GridSize/2,GridSize/2].real*np.sqrt(2)
deltakf[GridSize/2,GridSize/2,0] = deltakf[GridSize/2,GridSize/2,0].real*np.sqrt(2)
deltakf[GridSize/2,0,GridSize/2] = deltakf[GridSize/2,0,GridSize/2].real*np.sqrt(2)
deltakf[GridSize/2,GridSize/2,GridSize/2] = deltakf[GridSize/2,GridSize/2,GridSize/2].real*np.sqrt(2)
deltakf[0,0,GridSize/2+1:] = deltakf[0,0,1:GridSize/2][::-1].conjugate()
deltakf[0,GridSize/2+1:,0] = deltakf[0,1:GridSize/2,0][::-1].conjugate()
deltakf[GridSize/2+1:,0,0] = deltakf[1:GridSize/2,0,0][::-1].conjugate()
deltakf[0,1:GridSize,GridSize/2+1:] = deltakf[0,1:GridSize,1:GridSize/2][::-1,::-1].conjugate()
deltakf[1:GridSize,0,GridSize/2+1:] = deltakf[1:GridSize,0,1:GridSize/2][::-1,::-1].conjugate()
deltakf[1:GridSize,GridSize/2+1:,0] = deltakf[1:GridSize,1:GridSize/2,0][::-1,::-1].conjugate()
deltakf[1:GridSize,GridSize/2+1:,GridSize/2] = deltakf[1:GridSize,1:GridSize/2,GridSize/2][::-1,::-1].conjugate()


deltak_half_processed = deltakf[:,:,0:GridSize/2+1]
```

```python
### inverse fourier transform
Time_half = np.zeros([100,1])
Time_half_processed = np.zeros([100,1])
Time_full = np.zeros([100,1])

for i in range(0,100):
    start = time.time()
    deltax_half = np.fft.irfftn(deltak_half)
    t1= time.time()
    Time_half[i] = t1-start
    deltax_half_processed = np.fft.irfftn(deltak_half_processed)
    t2=time.time()
    Time_half_processed[i] = t2-t1
    deltax = np.fft.ifftn(deltakf)
    t3=time.time()
    Time_full[i] = t3-t2
    
```

```python
Time_half_mean = np.mean(Time_half)
Time_half_processed_mean = np.mean(Time_half_processed)
Time_full_mean = np.mean(Time_full)


print Time_half_mean 
print Time_half_processed_mean 
print Time_full_mean 
```
